Computers generally store numerical data using binary
or base-2: sequences of \(0\)s and \(1\)s.
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
Just like base-10 numbers describe values in terms of powers of
ten, binary numbers describe values using powers of two.
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
 \[\begin{align*}125 \amp= 1\times 100 + 2\times 10 + 5\times 1\\\amp= 1\times 10^2 + 2\times 10^1+ 5\times 10^0\\\amp= 1\times 64 + 1\times 32 + 1\times 16 + 1\times 8 + 1\times 4 + 0\times 2 + 1\times 1\\\amp= 1\times 2^6 + 1\times 2^5 + 1\times 2^4 + 1\times 2^3 + 1\times 2^2 + 0\times 2^1 + 1\times 2^0\\\amp= 1111101_2\\\end{align*}\]
 </span>
 

</div>

 <h3>
 Activity 1.2.4</h3>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
While <tt>10^3</tt> in writing might seem to refer to the number \(10^3=1000\),
in many programming languages such as Python 3, the caret <tt>^</tt> instead refers to the
bitwise XOR operation,
which can be evaluated by hand using a variation of long addition on the
binary form of each number.
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
 \[10 \verb|^| 3 = 1010_2 \verb|^| 0011_2\]
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
 \[\begin{array}{r} 1010_2\\\verb|^|\hspace{1em} 0011_2\\\hline 1001_2 \end{array}\]
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
Therefore \(10\verb|^| 3=1001_2=9\).
 </span>
 

</div>
