<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
Markdown has many features for writing formatted documents using plain text, but
it does not provide much support for mathematical notation out-of-the-box.
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
Fortunately, many platforms that support Markdown also support LaTeX
(pronounced "lay-tek" or "lah-tek",
and stylized as \(\LaTeX{}\)) markup for mathematical notation.
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <div><b>a. </b></div>
 
 <span>
To denote part of your writing as mathematical notation in a Markdown cell,
use <tt>\(</tt> and <tt>\)</tt> to surround your mathematics.
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
Create two Markdown cells with the text <tt>The solution to x+1=7 is x=6.</tt>.
Then use <tt>\(</tt> and <tt>\)</tt> to surround the mathematical notation in one of those cells
to observe the difference. (Don't forget to use Shift+Enter to
display the formatted text.)
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <div><b>b. </b></div>
 
 <span>
Now create two Markdown cells with the text <tt>The solutions to x^2-1=0 are x=-1 and x=1.</tt>,
and use <tt>\(</tt> and <tt>\)</tt> to surround the mathematics in one cell.
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
Some LaTeX notation is rather straight-forward, with literal notation such as
<tt>+</tt> and <tt>-</tt>, and intuitive notation such as <tt>^</tt> for superscripts
and <tt>_</tt> for subscripts: <tt>a_0+a_1^2=-a_2^3</tt> produces \(a_0+a_1^2=-a_2^3\).
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
But multiplication cannot be represented by, say, <tt>x</tt>, since that represents the variable
\(x\). (Instead we write <tt>2 \times 3</tt> to produce \(2 \times 3\).)
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">

Dr. Burn's TeX Command Reference
and tools like
DeTeXify
allow for easy lookup of LaTeX commands as needed.

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>
LaTeX markup is used by many applications for typesetting mathematical notation.
But it's also possible to write entire documents in LaTeX, that can be distributed
as PDFs. Since it's very easy to write mathematics, manage bibliographies,
and integrate a LaTeX document with computer code, it's very common for STEM
researchers to author documents in LaTeX.
 </span>
 

</div>

<div class="not-editable" style=" background-color:#f8f8f8;padding:1em;border-radius:10px;box-shadow:4px 4px 3px #ddd;margin:5px; ">


 <span>

Try it out by following these steps.
 </span>

 <ul>
 <li>
 
 <span>
From the + New menu option, create a new LaTeX document in the
same folder as this notebook.
A default document will appear on the left, and its PDF output appears on the right.
 </span>
 
 </li>
 
 <li>
 
 <span>
The top part of the code is the preamble with various configurations,
most of which you don't need to worry about. There is a place to define the title
and author of the paper. Change the title of the paper to match your favorite TV show,
and change the author to be named after a character on that show.
 </span>
 
 </li>
 
 <li>
 
 <span>
By default the produced PDF will include today's date. This can be configured by
inserting the command <tt>\date{Month Day, Year}</tt> in the preamble. Use this
to change the date to be one week from today instead.
 </span>
 
 </li>
 
 <li>
 
 <span>
The body of the document is between the <tt>\begin{document}</tt>
and <tt>\end{document}</tt> commands. The <tt>\maketitle</tt> command tells LaTeX
